```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
x_train=pd.read_csv('/content/train.csv')
y_train=pd.read_csv('/content/trainLabels.csv')
x_test=pd.read_csv('/content/test.csv')
```

```python
x_train=np.asarray(x_train)
x_test=np.asarray(x_test)
y_train=np.asarray(y_train)
```

```python
print('training_x.shape',x_train.shape)
print('training_y.shape',y_train.shape)
print('testing_x.shape',x_test.shape)
```

```python
x_all=np.r_[x_train,x_test]
print(x_all.shape)
```

```python
KN=KNeighborsClassifier()
KN.fit(x_train,y_train)
```

```python
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
```

```python
#GridSearchCV
param_grid={'n_neighbors':np.arange(1,50)}
knn=KNeighborsClassifier()
knn_cv=GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(x_train,y_train)
```

```python
print( 'best estimator KNN:',knn_cv.best_estimator_,
      'Best Score', knn_cv.best_estimator_.score(x_train,y_train))
```

```python
#Gridsearchcv
grid_search_rf = GridSearchCV(RF, param_grid=dict( ), verbose=3,scoring='accuracy',cv=10).fit(x_train,y_train)
```

```python
print ('best estimator RandomForest:',grid_search_rf.best_estimator_,
       'Best Score', grid_search_rf.best_estimator_.score(x_train,y_train))
```

```python
rf_best = grid_search_rf.best_estimator_
rf_best.fit(x_train,y_train)
```
